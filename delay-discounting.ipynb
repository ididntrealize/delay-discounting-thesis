<h2>Abstract</h2>

<p>Impulsivity is the preference for short term rewards that may be at the expense of rewards in the long term. Delay Discounting is the reduction of the value of a reinforcer due to the delay required to receive it. Delay Discounting tasks have been designed to measure impulsivity in humans and non-human animals by offering choices between Smaller Sooner (SS) and Larger Later (LL) reinforcers. This study investigated the relationship between impulsivity as measured by a Delay Discounting procedure and reinforcer checking behavior. Subjects were 8 Wistar rats pre-trained in a Delay Discounting task consisting of 5 blocks of 8 trials, with each block increasing delay, 0 s, 4 s, 8 s, 16 s, and 32 s. In each trial, subjects chose between 1 45 mg pellet immediately (SS) and 3 45 mg pellets after the current block delay (LL). Given a LL choice, head entries to the food dish were recorded until the delivery of the delayed reinforcer. It was hypothesized that rats scoring higher in impulsivity would check for delayed reinforcers earlier and more often. No relationship found between impulsivity and time to first head entry. A significant correlation between impulsivity and number of head entries was found in the opposite direction of our hypothesis. Subjects who scored lower on our impulsivity measure checked for delayed reinforcers significantly more often in the 16 s and 32 s delay blocks.</p>

<h2>Introduction</h2>
<p>Impulsivity is the preference for immediate rewards in place of or at the expense of larger delayed rewards. Impulsive behavior has been strongly associated with gambling addiction (Alessi &amp; Petry, 2003), cigarette smoking (Bickel et. al, 1999; Kale et al., 2018), alcohol use (Granö et al., 2004), substance abuse (Kirby et al., 1999), risky drug behavior (Odum et al., 2000), and obesity (Rasmussen et al., 2010; Boomhower et al., 2013). Some studies have shown support for impulsivity as a causal factor of those problems (Granö et al., 2004). Other studies have suggested that impulsivity is a relatively enduring personality trait (Odum, 2011). Although many negative outcomes have been paired with impulsivity, it is not fully clear to what extent the negative outcomes encourage further impulsive choice. </p>
<p>Some genetic disorders are correlated with higher than normal impulsivity; for example children with attention-deficit/hyperactivity disorder (ADHD) discount future hypothetical rewards at a steeper rate than control participants (Wilson et al., 2010). Experiments have shown impulsivity moderation with dopamine reuptake inhibitors, like methylphenidate (Shiels &amp; Hawk, 2009). Interventions that reduce impulsive choice in at-risk individuals could be a key in preventing common precursors to early death: cigarette smoking, obesity, and alcohol/drugs. Some intervention strategies have been studied, for example time-based approaches and nature exposure. Time-based interventions, where rats are trained to act less impulsively by undergoing waiting times have been shown to reduce impulsive choice in rats (Stuebing et al., 2017). In humans, nature exposure interventions have reduced impulsivity (Berry et al., 2014). </p>
<p>Greater potential for some cognitive processes has also been indicated as a decreasing factor of impulsivity, for example IQ (Baker et al., 2003), Working Memory (Wesley &amp; Bickel, 2013), and verbal expression of reasoning. Benjamin et al. found that asking participants to explain impulsive choices reduced levels of impulsivity (2013). Students with lower levels of impulsivity have been shown to get better grades in university (Kirby et al., 2005). Succeeding in university is seen as a delayed reward, because the goal is often to improve one’s future at the expense of extra effort up front. Rather than try to fundamentally change impulsivity in people, some educators are moving towards a teaching model which takes impulsive tendency into account by offering more immediately relevant information, game-based learning, and open cooperation/competition with peers (Boiko, 2018). </p>
<h3>Socioeconomic Status and Impulsivity</h3>
	<p>Impulsive behavior has predicted the onset of behaviors like gambling in low but not high Socioeconomic Status (SES) participants (Auger et al., 2010). Other studies have shown that hardships found in low SES participants such as increased emotional reactivity and decreased working memory were associated with greater impulsive choice (Oshri et al., 2019; Hair et al., 2015). Studies showing that impulsivity can predict unfavorable life outcomes have faced criticism due to the relationship between low SES and impulsivity. Poverty alone is a reasonable explanation of substance and alcohol abuse (Swendsen &amp; Moal, 2011), but poverty is not sufficient to produce impulsive behavior. Jachimowicz et al. found that low income participants with a high degree of community trust make less impulsive choices than those without community trust. In this experiment, participants responded online to demographic, trust, and impulsive choice measures (2017). Individuals with impulse control disorders and impulsive related conditions have reported lower quality of life (Chamberlain &amp; Grant, 2019) and low quality of life of parents can create less emotionally and cognitively supportive home environments for low SES children (Eamon, 2002).
    </p>
    <p>
Trust in your direct environment has also been shown to have a large effect on impulsive choice. In one well-cited study, child participants were given the option of eating one marshmallow immediately, or two marshmallows after a delay (Mischel, 1970). A later experiment added a pre-test where participants were shown that their environment was either reliable or unreliable prior to taking the marshmallow test. Kidd et al. asked participants to complete an art project before doing the marshmallow test, with intentionally insufficient provided art materials. An experimenter told participants they would return with better art materials, then in the reliable condition, the experimenter would return with better crayons. In the unreliable condition, the experimenter returned and said they didn’t have any other crayons. Participants in the reliable condition waited on average three times longer for the second marshmallow, and it’s estimated that many participants in that condition would have waited even longer had the delay been larger than 15 minutes (2012).</p>
<h3>Measuring Impulsive Choice</h3>
<p>Impulsivity has been measured in humans and non-human animals using food reinforcers (Mischel, 1970; Stuebing et al., 2018). In humans, impulsivity can also be studied using hypothetical monetary rewards (Benjamin et al., 2013; Odum et al., 2000), actual monetary rewards (Shiels, 2009), and self-report measures (Oshri et al., 2019). Hypothetical rewards have been shown to work in some cases but not in others. One study by Shiels et al. found that children with ADHD choose hypothetical rewards differently than real rewards. In the experimental condition, impulsivity was reduced in child participants taking the ADHD medication methylphenidate only when real delays and rewards were used (2009). Studies using mixed real and hypothetical discounting within participants have been suggested to be unreliable (Shamosh &amp; Gray, 2007). </p>
<h3>Types of Delay Discounting Procedures</h3>
<p>Different procedures have been developed to systematically offer subjects and participants varieties of choices between reinforcer quantities and delay. One such procedure called titrating modulates the quantity of immediate rewards depending on Larger Later (LL) and Smaller Sooner (SS) choices in a discounting task. When a subject chooses LL, the quantity of the SS reinforcer goes up, and when the subject chooses SS, the quantity of the SS reinforcer goes down. For example, if the SS quantity starts at 3 sucrose pellets, a SS choice would decrease the subsequent trial SS quantity to 2 pellets but a LL choice would increase the subsequent trial SS quantity to 4. After a number of trials, indifference points can be observed for each subject when responses stabilize. Based on subjects’ final SS reinforcer quantity, relative levels of delay discounting can be compared (Odum et al., 2000; Du et al., 2002). 
</p><p>
Rodzon et al. compared titrating methods to fixed sequence procedures, finding that the 2 methods are more or less equivalent (2011). A fixed sequence/interval procedure offers participants choices in multiple series of ascending or descending delays (Madden et al., 1997). A common interval discounting procedure used to study impulsiveness in rats introduced by Evenden and Ryan (1996) has been extensively utilized in modern impulsivity research (Shiels et al., 2009; Boomhower et al., 2013; Cano et al., 2015; Kale et al., 2018; Chamberlain &amp; Grant, 2019). Evenden and Ryan measured impulsivity by offering food reinforcer choices to rats. Rats chose between a SS and LL reward at ascending fixed interval delays. In the first 8 trials, rats chose between a 1 pellet reinforcer and a 3-5 pellet reinforcer with 0 s of delay. Each successive block of 8 trials increased the delay to delivery of the larger 3-5 pellet reinforcer, resulting in 10 s delay, 20 s delay, 40 s delay, and 60 s delay. The first trial block with 0 s delay functioned as a baseline of reasonable responses from the rat, i.e., rats choosing 1 pellet when there was no delay to receive 3-5 pellets may be determined unfit for the experiment (1996). Prior to inclusion in the delay discounting task, researchers often include pre-training tasks to verify a subject is acting intentionally. For example, our pre training phases can be found in Figure 1, and were based off of Prince et. al (2020) and Evenden &amp; Ryan’s methods (1996).</p>
<h3>Current Study Research Design Choices</h3>
<p>The current study will test levels of impulsivity in rats using the methods of Evenden and Ryan (1996) with the added dimension of reinforcer delivery checking behavior. A pilot study in our lab using this research design found a correlation between impulsivity and the first instance of checking behavior. An insignificant effect was found between impulsivity and number of checking behaviors which was in the opposite of the direction that we hypothesized, suggesting that we repeat our experiment. The current study used different delay block times than Evenden and Ryan. Instead of delay blocks of 10 s, 20 s, 40 s, and 60 s, delay blocks were based off of Maguire et al., 4 s, 8 s, 16 s, and 32 s (2014). The choice to use shorter delay blocks was due to subjects never choosing LL in the 40 s and 60 s blocks during our pilot study. Impulsivity was measured by calculating the Area Under the Curve (AUC) of discounting curves and checking behavior was defined as inserting head into the Skinner Box food dish. We hypothesized that rats with higher in impulsivity will exhibit more checking behavior per LL choice. Additionally we hypothesized that rats with higher impulsivity will exhibit checking behavior earlier after LL choices.</p>


<h2>Methods</h2>
<h3>Subjects</h3>
	<p>Ten female Wistar rats (Charles River, Wilmington, MA) of approximately 1 year of age at the start of experimentation were used as subjects. Prior to inclusion in experimentation, rats were taught basic lever training procedures in the University of Alaska’s Psychology 468: Learning & Behavior class. Two subjects were excluded over the course of the experiment. One subject was euthanized due to a tumor and another was removed due to slow progress in phase 2 and related time constraints. Subjects were housed in dual plastic 41 cm × 21 cm × 20 cm cages (Allentown Caging, Allentown, NJ) connected by 7.75 cm diameter, 35 cm long sections of acrylonitrile butadiene styrene (ABS) pipe. Subjects were housed communally in groups ranging from 2 to 4 rats per fixture. Enclosures had pine shavings as bedding, were kept in a temperature controlled room, and were kept on a 12 hour light cycle starting at 19:00 and going dark at 07:00 every day. Subjects were fed ad libitum Mazuri® Rodent Pellets (PMI Nutrition International, LLC, Brentwood, MO) and had ad libitum access to tap water. Subjects were put into Skinner Box sessions between 11:00 and 17:00 during the experiment Sunday through Friday. These experiment subject conditions were approved by the university Institutional Animal Care and Use Committee (UAA IACUC Protocol# 1966573).
</p>
<h3>Apparatus</h3>
	<p>The current research utilized four  31 cm × 26 cm × 23 cm modularized Skinner Boxes constructed of components made by Med Associates (St. Albans, VT). Modules used in our Skinner Boxes include retractable levers (ENV-112CM), one 100 mA incandescent white stimulus light (ENV-221M) 4 cm above each lever, a 5 cm × 5 cm × 2.5 cm food receptacle (located 2 cm from Skinner Box floor), a Med Associates pellet dispenser (ENV-203-45), and a Med Associates head entry detector (ENV-254-CB). Pellet dispensers dispensed a mix of 45 mg precision pellets of the flavors, grain, banana, sucrose, and chocolate (Bio-Serv®, Frenchtown, NJ). All the Med Associates Skinner Boxes were controlled from a PC using Med-PC IV software. </p>
<h3>Procedure</h3>
	<p>Subjects were required to pass two preliminary phases prior to eligibility to be included in our third phase which produced the data analyzed in this experiment. Details about the procedures and completion criteria can be found in Figure 1. In phase 1, subjects were placed into Skinner Boxes and a program was activated which permanently extended a lever. Lever presses in phase 1 always resulted in delivery of one reinforcer. Criteria for completion in this phase consisted of 50 or more lever presses in a session of 20 minute duration. This phase was repeated for both left and right levers until subjects completed 50 or more lever presses in a single session for the left lever and another for the right lever.</p>
	<p>In phase 2, subjects were put into Skinner Boxes and a program was activated which extended both levers simultaneously. The purpose of this phase was to ensure that subjects most often chose a lever which resulted in a larger quantity of reinforcers. In this program, both levers extend every 90 seconds and subjects could choose to press either the left or the right lever, causing both levers to retract. Levers automatically retracted after 20 seconds without a choice, and the program concluded after 60 minutes for a total of 40 trials. During this phase, subjects were counterbalanced into two equal groups. One group received three reinforcers for a left lever selection and only one reinforcer for a right lever selection. The other group was identical except that the right lever press resulted in three reinforcers and the left lever press resulted in one reinforcer. Both lever presses in both groups resulted in immediate delivery of reinforcers. The completion criteria for this phase was two consecutive program sessions where a subject chose their respective larger lever for 80% of total choices or greater.</p>
	<p>The final phase doubled as an assurance of stable responses and the data collection phase. During this phase, subjects were placed in Skinner Boxes and a program was started that incrementally increased the amount of delay to the larger lever reinforcer in 5 blocks of 8 trials each, with the delays per block of 0 s, 4 s, 8 s, 16 s, and 32 s. Both levers were extended every 90 seconds, and both levers retracted automatically if neither was pressed in 20 seconds. The larger lever delivered reinforcers after a delay in this phase in all blocks except the first, so it became the Larger Later lever. After pressing LL, the latencies of subject head entry into the Skinner Box food dish were recorded using a photobeam head entry detector. On delivery of the LL reinforcer, the head entry timer was reset.</p>
	<p>In phase 3, each subject underwent 10 initial stability sessions, where the percentage of LL presses was recorded in each delay block. Stability ranges were calculated by setting the minimum and maximum percentage of LL choices per block for each of 10 stability trials. Any session where a subject’s percentage LL in any given block extended past their current stability range extended their stability range, but did not count towards the completion criteria for this phase. Completion criteria for phase 3 was 5 consecutive sessions where subject percentage LL fell within the subject’s stability range for each delay block. Data analysis was then performed on the 5 sessions within the subject stability range.
</p>

<h2>Results</h2>
<p>
    Overall, our subjects' data conformed to a discounting curve of descending percentage of LL choices as delay increased (See Figure 2). The average percentage of LL choices was 90.125 (SD = 0.09) in the 0 s block, 86.50 (SD = 0.15) in the 4 s block, 60.38 (SD = 0.30) in the 8 s block, 54.00 (SD = 0.34) in the 16 s block, and 42.63 (SD = 0.34) in the 32 s block. We used Area Under the Curve (AUC) to score each subject on impulsivity. Myerson et al. first developed AUC as a univariate measure of discounting which does not require fitting data to a line of best fit (2001). AUC scores per subject can be found in Table 1, and ranged from .33 to .92 (SD = 0.23).</p>
    <p>
Mean head entries per block are graphed in Figure 3. On average, subjects made 4.45 head entries in the 4 s delay block (SD = 2.65), 7.0 head entries in the 8 s delay block (SD = 4.77), 6.54 head entries in the 16 s delay block (SD = 4.35), and 6.45 head entries in the 32 s delay block (SD = 3.32). A one-way repeated measures ANOVA was conducted to examine the effect of delay block influence on number of head entries, F(3, 21) = 1.62, p = .22</p>
<p>
The latency from LL choice to first head entry in the Skinner Box food dish is shown across delay blocks in Figure 4. On average, subjects first checked for reinforcer delivery at .7 seconds in the 4 s delay block (SD = 0.23), at 1.35 seconds in the 8 s delay block (SD = 0.73), at 2.3 seconds in the 16 s delay block (SD = 1.37), and at 2.83 seconds in the 32 s delay block (SD = 1.29). A one-way repeated measures ANOVA was conducted to differentiate delay block influence on latency of first head entry, F(3, 21) = 8.25, p &lt; .001. Six paired sample t-tests were done to find differences between delay blocks, because 6 tests were used, a Bonferroni test was used to reduce the alpha value to .008. A significant effect was found between the 4 s block and the 32 s block t(7) = -4.76, p = .002. No other significant differences were found using the reduced alpha, .008.</p>
    <p>
The first hypothesis was that subjects with low AUC would exhibit more checking behavior for delayed reinforcers. Correlation analysis was done for AUC and number of head entries for every delay block (see Figure 5). Trends were found in the opposite direction as our hypothesis, but without statistical significance in the 4 s block, r(6) = 0.26, p = .54 and in the 8 s block, r(6) = 0.11, p = .79. In the 2 higher delay blocks, however, significant effects were found. In the 16 s block, r(6) = 0.759, p = .029 and in the 32 s block, r(6) = 0.755, p = .03. Therefore in the 2 longer delay blocks, higher AUC was significantly positively correlated with number of head entries.</p>

<p>The second hypothesis was that subjects with low AUC would check for delayed reinforcers sooner after making a LL choice. No correlations were found in any of the delay blocks (see Figure 6). In the 4 s block, r(6) = 0.20, p = .63, in the 8 s block, r(6) = 0.03, p = .94, in the 16 s block, r(6) = 0.10, p = .82, and in the 32 s block, r(6) = 0.21, p = .63.
</p>

<h2>Discussion</h2>

<p>This research study examined the relationship between impulsivity and delayed reinforcer checking behavior in rats via head entries into a food dish prior to delivery of a reinforcer. It was hypothesized that rats scoring higher on our impulsivity measure would check for delayed reinforcers sooner and more often. Results suggested that the opposite was the case for the number of checking behaviors in the 16 s and 32 s delay blocks. Time to first head entry was unrelated to subject scores from our impulsivity measure.</p>
<p>
Impulsivity and impatience are often used interchangeably (Lempert &amp; Phelps, 2016), so it was expected that more impulsive subjects would exhibit more reinforcer checking behavior. One possible explanation for the findings of this experiment is that more patient rats would have a better expectation that food would be delivered, and hence look for it more often. It would make sense that impulsive rats may not have checked as much for rewards they may not have expected to arrive. If checking behavior was a result of better working memory representations and better working memory was associated with less impulsivity, it would make sense that less impulsive subjects checked more often for delayed reinforcer delivery. Some research in human participants has shown that intelligence and working memory are negatively related to impulsivity (Baker et al., 2003; Wesley &amp; Bickel, 2013). However, an experiment by Dellu-Hagedorn found that working memory was not associated with Delay Discounting impulsivity in rats (2006).
    </p>
    <p>
In the famous Mischel &amp; Ebbesen marshmallow test, child participants developed elaborate self distraction techniques to improve their ability to wait for a Larger Later reward. The researchers suggested that the ability to divert one’s attention may be a key step to successfully delaying gratification (1970). An assumption in the marshmallow experiment (and in the official definition of the word patience) is that patience involves tolerating delay well. In the marshmallow experiment, diverting attention contributed to not consuming the constantly available marshmallow too early to receive the reward for waiting. Because diverting attention could not affect delivery of a LL reinforcer, food attention related behaviors such as delayed reinforcer checking did not result in fewer reinforcers. 
    </p>
    <p>
Another possibility is that our impulsivity measure was confounded by the subjective value of our reinforcer. Subjects that may have acted more impulsively otherwise, but valued the our selected reinforcers higher may have chosen Larger Later due to preference, not patience. Preferential and larger reinforcers resulting in less steep discounting curves is called the magnitude effect. This effect has been well documented in human subjects (Baker et al., 2003). In rat subjects, some studies have shown no difference in discounting curve steepness between most and least preferred reinforcer flavors (Calvert et al., 2010). However, a study by Prince et. al. showed the magnitude effect in rat subjects using food restriction conditions rather than size to increase the subjective value of reinforcer (2020).
One strength of the current experiment was that it built off of a well defined measure of impulsivity, Delay Discounting. We added the novel factor of delayed reinforcer checking behavior, but no new subject behavior was required nor were any intrusions made to the existing procedural experience of subjects. A limitation of the current study was sample size. Future research could benefit greatly from larger sample sizes. Another possible limitation was that all subjects received the same mix of reinforcer flavors. While this avoided using one flavor that may have differed greatly in subject preference, a better controlled alternative would have been to perform flavor preference tests across subjects so reinforcer preference could have been equated. An alternative method to using preferred flavors would be to restrict subject food intake, and see if we still find a correlation between head entries and impulsivity. </p>
    <p>
Going forward, more research could be done to untangle the concepts of patience and impulsivity. For example, a method similar to the marshmallow experiment could be developed to train rat subjects to ignore a smaller reinforcer constantly present in their environment in order to obtain a larger reinforcer if they can wait for a given delay. One study with crow and raven subjects used such a procedure, finding that subjects would exchange currently available reinforcers for more preferred reinforcers after various delays (Hillemann et al., 2014). The data from a self-control specific measure like that could then be compared with an impulsivity measure like Delay Discounting. 
    </p>
    <p>
If the results of the current study are a valid representation of impulsivity and patience, then the language that researchers use to describe non-impulsive subjects may have to change. Researchers may also have to adjust their assumptions about the implications of impulsivity research if patience and impulsivity are different traits. In conclusion, low impulsivity and delayed reinforcer checking behavior were significantly positively correlated in rat subjects performing a Delay Discounting task. Since impatience and impulsivity are used interchangeably, more research is warranted to untangle where these traits overlap and where they differ.</p>
