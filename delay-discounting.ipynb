<center><strong>Abstract</strong></center>

<p>Delay Discounting is the negative relationship of delayed delivery on the value of a reinforcer. Steeper discounting of increasingly delayed reinforcers has predicted numerous unfavorable outcomes such as substance abuse, gambling, and binge eating. Delay Discounting tasks have been designed to measure impulsivity, where humans and non-human animals choose between Smaller Sooner (SS) and Larger Later (LL) reinforcers. The current study will investigate patience as a dimension of LL choices by tracking checking behavior prior to reinforcer delivery. Subjects will be 10 Wistar rats pre-trained in a Delay Discounting task using food reinforcers and placed on an ad libitum diet with access to sweet food reinforcers only during hour long sessions in Skinner Boxes. No more than once per day, subjects will complete 5 blocks of 8 trials, with each block increasing delay, 0s, 4s, 8s, 16s, and 32s. SS and LL choices will be recorded, and given a LL choice, latency of any head entries to the food dish will also be recorded. A previous experiment showed a positive correlation between self-control (high AUC) and time until first head entry. We hypothesize that subjects with higher impulsivity will be lower in patience, checking for the LL reinforcer sooner and more often.</p>

<center><strong>Introduction</strong></center>

<p>
	Impulsivity is the preference for immediate rewards in place of or at the expense of larger delayed rewards. Impulsive behavior has been strongly associated with gambling addiction (Alessi &amp; Petry, 2003), cigarette smoking (Bickel et. al, 1999; Kale et al., 2018), alcohol use (Granö et al., 2004), substance abuse (Kirby et al., 1999), risky drug behavior (Odum et al., 2000), and obesity (Rasmussen et al., 2010; Boomhower et al., 2013). Some studies have shown support for impulsivity as a causal factor of those problems (Granö et al., 2004), others have suggested impulsivity is a relatively enduring personality trait (Odum, 2011). Although many negative outcomes have been paired with impulsivity, it is not fully clear to what extent the negative outcomes encourage further impulsive choice.</p>
<p>
	Some genetic disorders are correlated with higher than normal impulsivity; for example children with attention-deficit/hyperactivity disorder (ADHD) discount future hypothetical rewards at a steeper rate than control participants (Wilson et al., 2010). Experiments have shown impulsivity moderation with dopamine reuptake inhibitors, like methylphenidate (Shiels &amp; Hawk, 2009). Interventions that reduce impulsive choice in at-risk individuals could be a key in preventing common precursors to early death: cigarette smoking, obesity, and alcohol/drugs. Some intervention strategies have been studied, for example time-based approaches and nature exposure. Time-based interventions, where rats are trained to act less impulsively by undergoing waiting times have been shown to reduce impulsive choice in rats (Stuebing et al., 2017). In humans, nature exposure interventions have reduced impulsivity (Berry et al., 2014).</p>
<p>
	Greater potential for some cognitive processes has also been indicated as a decreasing factor of impulsivity, for example IQ (Baker et al., 2003), Working Memory (Wesley &amp; Bickel, 2013), and verbal expression of reasoning. Benjamin et al. found that asking participants to explain impulsive choices reduced levels of impulsivity (2013). Students with lower levels of impulsivity have been shown to get better grades in University (Kirby et al., 2005). Succeeding in university is seen as a delayed reward, because the goal is often to improve one’s future at the expense of extra effort up front. Rather than try to fundamentally change impulsivity in people, some educators are moving towards a teaching model which takes impulsive tendency into account by offering more immediately relevant information, game-based learning, and open cooperation/competition with peers (Boiko, 2018).</p>
<br><strong>Socioeconomic Status and Impulsivity</strong>
<p>
	Impulsive behavior has predicted the onset of behaviors like gambling in low but not high Socioeconomic Status (SES) participants (Auger et al., 2010). Other studies have shown that hardships found in low SES participants such as reduced emotional reactivity and working memory were associated with impulsive choice (Oshri et al., 2019; Hair et al., 2015). Studies showing that impulsivity can predict unfavorable life outcomes have faced criticism due to the relationship between low SES and impulsivity. Poverty alone is a reasonable explanation of substance and alcohol abuse (Swendsen &amp; Moal, 2011), but poverty is not sufficient to produce impulsive behavior. Jachimowicz et al. found that low income participants with a high degree of community trust make less impulsive choices than those without community trust (2017). Individuals with impulse control disorders and impulsive related conditions have reported lower quality of life (Chamberlain &amp; Grant, 2019) and low quality of life of parents can create less emotionally and cognitively supportive home environments for low SES children (Eamon, 2002).</p>
<p>
	Trust in your direct environment has also been shown to have a large effect on impulsive choice. In one well-cited study, child participants were given the option of eating one marshmallow immediately, or two marshmallows after a delay (Mischel, 1970). A later experiment added a pre-test where participants were shown that their environment was either reliable or unreliable prior to taking the marshmallow test. Kidd et al. asked participants to complete an art project before doing the marshmallow test, with intentionally insufficient provided art materials. An experimenter told participants they would return with better art materials, then in the reliable condition, the experimenter would return with better crayons. In the unreliable condition, the experimenter returned and said they didn’t have any other crayons. Participants in the reliable condition waited on average three times longer for the second marshmallow, and it’s estimated that many participants in that condition would have waited even longer had the delay been larger than 15 minutes (2012).</p>
<br><strong>Measuring Impulsive Choice</strong>
<p>
	Impulsivity has been measured in humans and non-human animals using food reinforcers (Mischel, 1970; Stuebing et al., 2018). In humans, impulsivity can also be studied using hypothetical monetary rewards (Benjamin et al., 2013; Odum et al., 2000), actual monetary rewards (Shiels, 2009), and self-report measures (Oshri et al., 2019). Hypothetical rewards have been shown to work in some cases but not in others. One study by Shiels et al. found that children with ADHD choose hypothetical rewards differently than real rewards. In the experimental condition, impulsivity was reduced in child participants taking the ADHD medication methylphenidate only when real delays and rewards were used (2009). Studies using mixed real and hypothetical discounting within participants have been suggested to be unreliable (Shamosh &amp; Gray, 2007).</p>
<br><strong>Types of Delay Discounting Procedures</strong>
<p>
	Different procedures have been developed to systematically offer subjects and participants varieties of choices between reinforcer quantities and delay. One such procedure called titrating modulates the quantity of immediate rewards depending on Larger Later (LL) and Smaller Sooner (SS) choices in a discounting task. When a subject chooses LL, the quantity of the SS reinforcer goes up, and when the subject chooses SS, the quantity of the SS reinforcer goes down. For example, if the SS quantity starts at 3 sucrose pellets, a SS choice would decrease the subsequent trial SS quantity to 2 pellets but a LL choice would increase the subsequent trial SS quantity to 4. After a number of trials, indifference points can be observed for each subject when responses stabilize. Based on subjects’ final SS reinforcer quantity, relative levels of delay discounting can be compared (Odum et al., 2000; Du et al., 2002).</p>
<p>
	Rodzon et al. compared titrating methods to fixed sequence procedures, finding that the 2 methods are more or less equivalent (2011). A fixed sequence/interval procedure offers participants choices in multiple series of ascending or descending delays (Madden et al., 1997). A common interval discounting procedure used to study impulsiveness in rats introduced by Evenden and Ryan (1996) has been extensively utilized in modern impulsivity research (Shiels et al., 2009; Boomhower et al., 2013; Cano et al., 2015; Kale et al., 2018; Chamberlain &amp; Grant, 2019). Evenden and Ryan measured impulsivity by offering food reinforcer choices to rats. Rats chose between a SS and LL reward at ascending fixed interval delays. In the first 8 trials, rats chose between a 1 pellet reinforcer and a 3-5 pellet reinforcer with 0s of delay. Each successive block of 8 trials increased the delay to delivery of the larger 3-5 pellet reinforcer, resulting in 10s delay, 20s delay, 40s delay, and 60s delay. The first trial with 0s delay functioned as a baseline of reasonable responses from the rat, i.e., rats choosing 1 pellet when there was no delay to receive 3-5 pellets may be determined unfit for the experiment (1996). Prior to inclusion in the delay discounting task, researchers often include pre-training tasks to verify a subject is acting intentionally. For example, our pretraining phases can be found in Figure 1, and were based off of Prince et. al (2020) and Evenden &amp; Ryan’s methods (1996).</p>
<br><strong>Current Study Research Design Choices</strong>
<p>
	The current study will test levels of impulsivity in rats using the methods of Evenden and Ryan (1996) with the added dimension of reinforcer delivery checking behavior. A pilot study in our lab using this research design found a correlation between impulsivity and the first instance of checking behavior. An insignificant effect was found between impulsivity and number of checking behaviors which was in the opposite of the direction that we hypothesized, suggesting that we repeat our experiment. Impulsivity was measured by calculating the Area Under the Curve (AUC) of discounting curves and checking behavior was defined as inserting head into the Skinner Box food dish. We hypothesized that rats with lower AUC will exhibit more checking behavior per LL choice. Additionally we hypothesized that rats with lower AUC will exhibit checking behavior earlier after LL choices.</p>
